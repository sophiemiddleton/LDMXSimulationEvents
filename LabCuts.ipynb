```python
# for Theta > 0.09, E > 1.2
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('LabPhotonAngleFF3.csv')

data.hist(bins=100)
plt.ylabel('N Events (Unweighted)')
plt.xlabel('Theta in Lab Frame [rad]')
plt.title('Binned Theta in Lab Frame for FFopt3')

plt.show()
```
